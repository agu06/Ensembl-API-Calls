```python
import requests

# URL for Ensembl species endpoint
url = "https://rest.ensembl.org/info/species?content-type=application/json"

# Send a GET request to the endpoint
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    species_list = data['species']

    # Print the species names
    print("List of species in Ensembl:")
    for species in species_list:
        print(species['name'])
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
import requests

# Base URL for Ensembl REST API
base_url = "https://rest.ensembl.org"

# Get the list of species
species_url = f"{base_url}/info/species?content-type=application/json"
species_response = requests.get(species_url)

if species_response.status_code == 200:
    species_list = species_response.json()['species']

    # Iterate through each species and fetch detailed information
    for species in species_list:
        species_name = species['name']
        species_id = species['id']

        # Fetch detailed information for each species
        detail_url = f"{base_url}/info/species/{species_id}?content-type=application/json"
        detail_response = requests.get(detail_url)

        if detail_response.status_code == 200:
            details = detail_response.json()
            # Print the detailed information
            print(f"Species: {species_name}")
            print(f"Scientific Name: {details.get('scientific_name')}")
            print(f"Common Name: {details.get('common_name')}")
            print(f"Assembly: {details.get('assembly')}")
            print(f"Taxonomy ID: {details.get('taxonomy_id')}")
            print(f"Other Info: {details}")  # You can customize this based on the structure of the details
            print("-" * 50)
        else:
            print(f"Failed to retrieve details for {species_name}: {detail_response.status_code}")
else:
    print(f"Failed to retrieve species list: {species_response.status_code}")
```
